Preliminary discussion of the implementation of two STS-NIR devices to collect spectral data. 

# Methods.

The setup detailed [here](https://github.com/calugo/AgroSightSPECY) was deployed to collect data and compute the reflectance index $R$. The system consists of two STS-NIR devices, a white reference and a cos-corrected filter respectively. Data for black current references was obtained by computing spectra by night with blocked fiber optics. The white references were collected by using a reflectant standard surface for the cos-corrected fiber and sunlight for the white reference filtered fiber. In both cases the measurements were obtained at the same time.

After a number of measurments the white surface standard was replaced by a green target. Three targets are reported here: venus flytrap, a single leaf plant and grass. while the white reference was kept recording.

### White references and Reflectance.

(__This needs to be agreed by Adolfo__)
A way of computing $R$ is somehow suggested [here](http://www.spectrecology.com/2015/08/two-channel-radiometry-for-reflectance-oceanview-schematic/), which is summarised as follows:

Let $y_{k}$, $k=1,2$ represent the raw spectra obtained with the STS-devices, 1 being the cosine-corrector fibre and 2 the white fiber and let $y_{k,b}$ represent the respective black references. Then, if we obtain a white reference $y_{k,w}$ for each device by means of a standard lambertian surface and __I supposse__ a maximum sunlight. Then $R$ can be computed as follows:

$$
\begin{array}{lcr}
\mbox{Substract black ref} & s_k=y_k-y_{k,b} & (1) \\
\mbox{Divide by white ref} & s'_k=s_k/y_{k,w} & (2) \\
\mbox{Divide cos-corr by white signal} & R=s'_1/s'_2 & (3) \\
\end{array} 
$$

Or:

$$R=\frac{y_{2,w}}{y_{1,w}} \frac{y_1-y_{1,b}}{y_2 - y_{2,b}}$$

where 1 and 2 refer to the cos-corrected and white filters respectively, the prefactor $$\Gamma_{2,1}=\frac{y_{2,w}}{y_{1,w}}$$ is then a quantity that as well as the black references requires to be precomputed. __However is it really neccessary?. Even more, is this the way you though to compute $R$?, actually is it also necessary to substract the black levels to the white references?__ In any case, below I show you some results obtained by following this method.
